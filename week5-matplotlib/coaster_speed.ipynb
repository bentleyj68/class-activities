```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0, 130, 10)
danger_drop = [9,8,90,85,80,70,70,65,55,60,70,65,50]
rail_gun = [75,70,60,65,60,45,55,50,40,40,35,35,30]
```

```python
plt.plot(x_axis, danger_drop, color="red", label="Danger Drop")
plt.plot(x_axis, rail_gun, color="green", label="RailGun")
```

```python
d_handle, = plt.plot(x_axis, danger_drop, color="red", label="Danger Drop")
r_handle, = plt.plot(x_axis, rail_gun, color="green", label="RailGun")
```

```python
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
```

```python
plt.xlim(90,120)
plt.ylim(0,95)
```

```python
plt.legend(handles=[d_handle, r_handle], loc="best")

```

```python
d_handle, = plt.plot(x_axis, danger_drop, color="red", label="Danger Drop")
r_handle, = plt.plot(x_axis, rail_gun, color="green", label="RailGun")
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.xlim(0,120)
plt.ylim(0,95)
plt.legend(handles=[d_handle, r_handle], loc="best")
plt.show()
```
