```python
# Dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load crime data set into pandas
crime_data = pd.read_csv("Resources/crime_data.csv")
crime_data[:5]
```

```python
# Generate a scatter plot of violent crime rate versus year
x_values = crime_data['violent crime rate']
y_values = crime_data['year']
plt.scatter(x_values,y_values)
plt.xlabel('Violent Crime')
plt.ylabel('Year')
plt.show()
```

```python
# Perform a linear regression on violent crime rate versus year
x_values = crime_data['year']
y_values = crime_data['violent crime rate']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2002, 650),fontsize=15,color="red")
plt.xlabel('Year')
plt.ylabel('Violent Crime')
plt.show()
```
