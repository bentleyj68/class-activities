```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_csv = 'Resources/got.csv'
got_data = pd.read_csv(got_csv)
got_data.head()
```

```python
# Get attacker and defender data
attackers = got_data['attacker_king'].value_counts()
defenders = got_data['defender_king'].value_counts()

attackers
```

```python
# Get total battle data
battle_data = attackers.add(defenders, fill_value=0)
battle_data
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar", facecolor="red")
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
plt.xlabel("King")
plt.tight_layout()
plt.show()
```
