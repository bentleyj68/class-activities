```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
csv_path = "Resources/avg_rain_aus.csv"
rain_df = pd.read_csv(csv_path)
rain_df.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(rain_df['City'], rain_df['Milimeters'], color='r', alpha=0.5, align="center")
tick_locations = [value for value in rain_df['City']]
plt.xticks(tick_locations, rain_df['City'], rotation='vertical')
plt.xlim(-1, len(rain_df['City']))
plt.ylim(0, max(rain_df["Milimeters"])+100)
plt.title("Average Rain per City")
plt.xlabel("City")
plt.ylabel("Average Amount of Rainfall in Milimeters")
plt.tight_layout()
```
