```python
#Dependencies
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url=' http://api.tvmaze.com/singlesearch/shows?q='
```

```python
#list of tv show titles to query
tv_shows = ["Altered Carbon", "Grey's Anatomy", "This is Us", "The Flash", "Vikings", "Shameless", "Arrow", "Peaky Blinders", "Dirk Gently"]

titles = []
rating = []

# make iterative requests to TVmaze search endpoint
for show in tv_shows:
    response = requests.get(url + show).json()
    titles.append(response['name'])
    rating.append(response['rating']['average'])
```

```python
# create dataframe
show_dict = {
    "Name": tv_shows,
    "Rating": rating
}

show_data = pd.DataFrame(show_dict)
show_data.head()
```

```python
# use pandas to create a bar chart from the dataframe
show_data.plot('Name', 'Rating', kind='bar')
```
